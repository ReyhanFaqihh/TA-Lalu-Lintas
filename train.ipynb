```python
import glob
import torch
from ultralytics import YOLO

# Cek apakah menggunakan GPU atau CPU
device = "cuda" if torch.cuda.is_available() else "cpu"

# Parameter Training
epochs = 50  # Jumlah epoch
batch_size = 16  # Batch size
img_size = 640  # Ukuran input YOLO

# Cari semua file YAML yang sesuai dengan nama "data_fold*.yaml"
yaml_files = sorted(glob.glob("data_fold*.yaml"))  # Ambil semua fold YAML

# Looping untuk setiap fold YAML
for yaml_file in yaml_files:
    fold_number = yaml_file.split("data_fold")[-1].split(".yaml")[0]  # Ambil nomor fold
    print(f"\n🚀 Training model untuk Fold {fold_number} dengan dataset: {yaml_file}")

    # Load model YOLOv11
    model = YOLO("yolo11n.pt")  # Gantilah dengan model awal yang digunakan

    # Jalankan Training
    model.train(
        data=yaml_file,
        epochs=epochs,
        batch=batch_size,
        imgsz=img_size,
        device=device,
        project="runs/detect",
        name=f"train_fold{fold_number}"  # Simpan model di folder khusus setiap fold
    )

    print(f"✅ Training selesai untuk Fold {fold_number}, model disimpan di runs/detect/train_fold{fold_number}/")

print("\n🎉 Semua training selesai!")
```

```python
import pandas as pd
import glob

# Cari semua file CSV hasil training dari setiap fold
csv_files = glob.glob("runs/detect/train_fold*/results.csv")

# Simpan hasil dalam list
results = []

for file in csv_files:
    df = pd.read_csv(file)
    last_epoch = df.iloc[-1]  # Ambil hasil dari epoch terakhir untuk setiap fold
    results.append(last_epoch)

# Gabungkan hasil ke dalam DataFrame
df_results = pd.DataFrame(results)

# Hitung rata-rata dan standar deviasi metrik
mean_results = df_results.mean()
std_results = df_results.std()

print("Rata-rata hasil training dari semua fold:")
print(mean_results)

print("\nStandar deviasi dari semua fold:")
print(std_results)
```
